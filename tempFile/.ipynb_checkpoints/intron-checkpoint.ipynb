```python
from BCBio.GFF import GFFExaminer
import os
import pandas as pd
import re
```

```python
base_path = "/Users/richard/local_data/gff3_Fall2014_intron_added/"
filenames = os.listdir(base_path)
```

```python
num_gene = []
num_exon = []
num_intron = []
print "filename", "#gene", "#exon", "#intron"
for filename in filenames:
    examiner = GFFExaminer()
    in_handle = open(base_path+filename)
    test = examiner.available_limits(in_handle)
    
    
    try:
        print filename[13:21], test['gff_type'][('gene',)], test['gff_type'][('exon',)], test['gff_type'][('intron',)]
    except:
        print
        print "problem with: "+filename
        print
        continue
    num_gene.append(test['gff_type'][('gene',)])
    num_exon.append(test['gff_type'][('exon',)])
    num_intron.append(test['gff_type'][('intron',)])
    in_handle.close()
```

```python
statistics = pd.DataFrame({"gene_count":num_gene, "exon_count":num_exon, "intron_count":num_intron}, 
             columns=["gene_count", "exon_count", "intron_count"])
```

```python
statistics.to_csv("exon_intron_count.csv")
```

```python
## make concise filename
filenames_short = []
for filename in filenames:
    match_obj = re.match("intron_added_(.*?).filtered_proteins",filename)
    filenames_short.append(match_obj.group(1))

with open("filenames_short.csv","w") as f:
    for name in filenames_short:
        f.write(name+"\n")
```

```python
full_table_names = pd.read_csv("name_exon_intron_count.csv")['name'] # name needed to add into the table
```

```python
# match rna seq count to exon, intron density
long_short_name_pair = pd.read_csv("../July29/long_short_name_pair_2015.csv")
short_long_dict = {long_short_name_pair['shortName'][i]:long_short_name_pair['longName'][i] 
                   for i in range(long_short_name_pair.shape[0])}

long_rna_pair =  pd.read_csv("../July29/rna_seq_count.csv")
long_count_dict = {long_rna_pair['name'][i]:long_rna_pair[' rna_seq_count'][i] 
                   for i in range(long_rna_pair.shape[0])}
```

```python
longname = []
shortname = []
counts = []
for name in full_table_names:
    try:
        long_name = short_long_dict[name]
        try:
            count = long_count_dict[long_name]
        except:
            count = 0
        longname.append(long_name)
        shortname.append(name)
        counts.append(count)
            
    except:
        print "problem with", name
        longname.append("N/A")
        shortname.append(name)
        counts.append(0)

rna_seq_intron_density = pd.DataFrame({'longName':longname,'shortName':shortname,'count':counts})
rna_seq_intron_density.to_csv("intron_density_rna_seq_count.csv", index = False)
```

```python
## filter table to exclude rna count zero species
data = pd.read_csv("./longNme_shortName_exon_intron_rna_seq_count.csv")
idx = data['rna_seq_count'] > 0
filtered_data = data[idx]
filtered_data.to_csv("longNme_shortName_exon_intron_rna_seq_count_filtered.csv",index = False)
```

```python
## final step, add gc
complete_data = pd.read_csv("../common_file/CompleteV6.csv")
```

```python
short_name_gc_dict = {complete_data['ShortName'][i]:complete_data['Genome_GC'][i] 
                      for i in range(complete_data.shape[0])}
```

```python
gc_list = []

for name in filtered_data['shortName']:
    try:
        gc = short_name_gc_dict[name]
        gc_list.append(gc)        
    except:
        print "problem with", name
        gc_list.append(-1)
```

```python
with open("gc_list.csv",'w') as f:
    for gc in gc_list:
        f.write(str(gc)+"\n")
```

```python
## final filtered table
filtered_table = pd.read_csv("longNme_shortName_exon_intron_rna_seq_count_filtered.csv")
idx = filtered_table['genome_gc'] > 0
final_table = filtered_table[idx]
final_table.to_csv("longNme_shortName_exon_intron_rna_seq_count_final.csv",index = False)
```
