```python
import pandas as pd
import urllib2
import re
import time
import numpy as np
```

```python
data = pd.read_csv("genome_file_info.csv")
names = data['name']
```

```python
unique_names = set(map(lambda x: " ".join(x.split()[:2]),names))
```

```python
base = "http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db={}&term={}"
db = "SRA"

def make_format(name): return "+".join(name.split())+"+"+"RNA"

def get_result_count(response):
    res = response.read()
    try:
        if "<Field>Organism</Field>" in res:
            res = res.split("<Field>Organism</Field>")
            res = res[1]
            return int(re.findall(r"<Count>(\d*?)</Count>", res)[0])
    except:
        print res
        return 0
        
    
# base.format(db,make_format(names[0])) #test code
```

```python
result_count = []
counter = 0
# for name in names:
for name in unique_names:
    counter += 1
    try:
        response = urllib2.urlopen(base.format(db,make_format(name)))
#         print make_format(name)
    except:
        response = None
#     responses.append(response)
    count = get_result_count(response)
#     print counter, (name, count)
    if count > 0:
        result_count.append((name, count))
        print counter, (name, count)
#     print response
```

```python
result_count = sorted(result_count)
```

```python
len(result_count)
```

```python
with open("rna_seq_count.csv","w") as f:
    f.write("name, rna_seq_count\n")
    for (x,y) in result_count:
        f.write(x+", "+str(y)+"\n")
```
