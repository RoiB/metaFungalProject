```python
'''
make query to ncbi regarding the 363 genomes to see if they have rna seq data 
'''
import pandas as pd
import urllib2
import xml.etree.ElementTree as ET
info = pd.read_csv("3_completeFileInfoWithStatistics.csv")
# info
```

```python
base = 'http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=SRA&term=biomol+rna[Properties]+AND+'
rnaSeqCount = []
for i in range(info.shape[0]):
    # long name splitting is important, I choose only first two
    longName = "+".join(info.loc[i]['longName'].split()[:2]) 
    try:
        response = urllib2.urlopen(base+longName)
        root = ET.fromstring(response.read())
        rnaSeqCount.append(int(root.find('Count').text))
    except:
        rnaSeqCount.append(0)
    print int(root.find('Count').text), info.loc[i]['longName']
```

```python
info['rnaSeqCount'] = rnaSeqCount
info.to_csv('completeInfoStatisticsRNASeq.csv', index = None)
```
