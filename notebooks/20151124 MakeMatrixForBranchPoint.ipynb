```python
## get bp prob matrix, address the issue of weak signal
for shortName in info.shortName:
    try:
        freqs = pd.read_csv('/home/richard/research/tests/Nov24BranchPoint/richard-exp-w9/{}_intron.fasta.mfout'.format(shortName), skiprows = 2, sep = '\t', header = None)
        mat = pd.DataFrame([[0]*9 for i in range(4)]); mat.index = 'T C A G'.split()

        seqs = list(SeqIO.parse('/home/richard/research/tests/Nov24BranchPoint/intron_long/{}_intron.fasta'.format(shortName),'fasta'))

        for i in range(freqs.shape[0]):
            motif = freqs[0][i]
#             print len(seqs[i])
            if 'N' not in motif and len(seqs[i])>500:
                for i in range(9):
                    mat.ix[motif[i]][i] += 1
        result = mat/freqs.shape[0]
        result.to_csv('/home/richard/research/tests/Nov24BranchPoint/bp_mat/{}.bp.mat'.format(shortName), header = None, sep = ' ')
    except:
        print 'problem with {}'.format(shortName)
```

```python
## bp to acc spacer length distribution
n = 45
target = '/home/richard/research/tests/Nov24BranchPoint/spacer_length_distr/'

for shortName in info.shortName:
    try:
        length_raw = pd.read_csv('/home/karl/richard-exp-w9/{}_intron.fasta.mfout'.format(shortName), skiprows = 2, sep = '\t', header = None)[1]

        lengths = {i:0 for i in range(n)} 
        for number in length_raw:
#             print number
            if number > 2:
                lengths[n -number]+=1
        total = sum(lengths.values())
        for i in range(n):
            lengths[i] = 1.*lengths[i]/total

        plt.figure()
        plt.plot(range(n), [lengths[i] for i in range(n)])
        plt.title('Branch Point to Acceptor spacer length distribution\n{}, shortName: {}\n GC: {}, Intron Density: {}'.format(info.ix[shortName]['longName'], shortName, info.ix[shortName]['gc'], info.ix[shortName]['inron density']))
        plt.xlabel('Length')
        plt.ylabel('Probability')
    #     plt.ylim(0,0.1)
        plt.savefig(target+shortName, bbox_inches='tight')
        plt.close()
    except:
        print 'problem with {}'.format(shortName)
```
