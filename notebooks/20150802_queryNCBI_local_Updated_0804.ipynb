```python
## new table for updated info
base = 'http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=SRA&term=biomol+rna[Properties]+AND+'
rnaSeqCount = []

for i in range(fullTable.shape[0]):
     
    try:
        # long name splitting is important, I choose only first two
        longName = "+".join(fullTable.loc[i]['longName'].split()[:2])
        
        response = urllib2.urlopen(base+longName)
        root = ET.fromstring(response.read())
        rnaSeqCount.append(int(root.find('Count').text))
    except:
        rnaSeqCount.append(0)
    print int(root.find('Count').text), fullTable.loc[i]['longName']
    
fullTable['rnaSeqCount'] = rnaSeqCount
fullTable.to_csv('fullTableInfoGff3GffRNA.csv', index = None)
```
